```python
data=yf.download(tick,start="2023-04-01",end="2024-03-31")
```

```python
data
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df2=df
```

```python
x=len(df)
arr=[0.0]*x
for index in range(0,9):
    arr[index]=df["Close"][index]
```

```python
sum2=0
for index1 in range(9,243):
    for index2 in range(index1-9,index1+1):
        sum2=sum2+df["Close"][index2]
    arr[index1]=sum2/10
    sum2=0
arr
```

```python
for index in range(0,len(arr)):
    arr[index]=round(arr[index],2)
    
```

```python
for index in df.index:
    df["Close"][index]=round(df["Close"][index],2)
```

```python
df["SMA_10"]=0
for index in range(0,243):
    df["SMA_10"][index]=arr[index]
df
```

```python
plt.figure(figsize=(8,6),dpi=150)
plt.title("TCS vs TCS SMA_10")
df["Close"].plot(label="TCS",color=["red","blue"])
df["SMA_10"].plot(label="TCS SMA_10 ")
plt.xlabel("Date")
```
