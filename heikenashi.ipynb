```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tick="WIPRO.NS"
```

```python
data=yf.download(tick,start="2023-04-01",end="2024-03-31")
df=pd.DataFrame(data)
df
```

```python
df2=pd.DataFrame()
```

```python
x=len(df)
```

```python
df2["Close"]=0.25*(df["Close"]+df["High"]+df["Low"]+df["Open"])
```

```python
df2["Open"]=0
for index in range(1,x):
    df2["Open"][index]=0.5*(df["Open"][index-1]+df["Close"][index-1])
```

```python
for index in range(0,x):
    df2["High"][index]=max(df2["Close"][index],df2["Open"][index])
    df2["High"][index]= max(df["High"][index],df2["High"][index])
def max(a,b):
    if(a>=b):
        return a
    else:
        return b
```

```python
list2=[df2["Open"],df2["Close"],df["Low"]]
df2["Low"]=np.min(list2)
```
