Final Capital

```python
initialcapital=100000
capital=initialcapital
for index in df.index:
    open=df["Open"][index]
    close=df["Close"][index]
    dailyreturns=(close/open-1)*100
    capital=capital*(1+dailyreturns/100)
    
finalcapital=capital   
print("Final Capital of Alice is :",round(finalcapital,2))       
```

Daily Returns

```python
sumdailyreturns=0
count=0
for index in df.index:
    open=df["Open"][index]
    close=df["Close"][index]
    dailyreturns=(close/open-1)*100
    sumdailyreturns=sumdailyreturns+dailyreturns
    count=count+1
    print(round(dailyreturns,2))
   
```

```python
df["Daily_returnspercentage"] = round((df["Close"]/df["Open"]-1)*100,2)
df
```

```python
x=len(df)
```

```python
df["Capital"]=0
```

```python
initialcapital=100000
capital=initialcapital
for index in df.index:
    df["Capital"][index]=capital*(df["Close"][index]/df["Open"][index])
    capital=capital*df["Close"][index]/df["Open"][index]
```

```python
df["Capital"]=round(df["Capital"],2)
df
```

Sharpe Ratio

```python
sumofsquares=0
mean=sumdailyreturns/x
for index in df.index:
    open=df["Open"][index]
    close=df["Close"][index]
    dailyreturns=(close/open-1)*100
    sumofsquares=sumofsquares+abs(mean-dailyreturns)*abs(mean-dailyreturns)
variance=sumofsquares/x
sd=(np.sqrt(variance))*np.sqrt(x)
sharperatio=(mean*x)/sd
print(round(sharperatio,2))
```

Another Method for Sharpe Ratio

```python
round(((df["Close"]/df["Open"]-1)*100).mean()/((df["Close"]/df["Open"]-1)*100).std()*np.sqrt(x),2)
```
